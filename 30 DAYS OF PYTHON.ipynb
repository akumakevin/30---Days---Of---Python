```python
# Day 1 – Introduction to Python
print("Hello, 30-Days-of-Python challenge!")

# Variables
integer_num = 10
float_num = 3.14
string_text = "Python is fun"
boolean_val = True

print(integer_num, type(integer_num))
print(float_num, type(float_num))
print(string_text, type(string_text))
print(boolean_val, type(boolean_val))

# Arithmetic operations
a, b = 7, 4
print(a + b, a - b, a * b, a / b, a // b, a % b, a ** b)

# String concatenation
first_name = "Kevin"
last_name = "Atandi"
full_name = first_name + " " + last_name
print(f"Full name: {full_name}")
```

```python
# Day 2 – Variables and Built-in Functions
x = 5
y = 3
name = "Kevin"

print(len(name))
print(type(x))
print(round(3.14159, 2))
print(abs(-10))

# Type conversion
num_str = "100"
num_int = int(num_str)
num_float = float(num_str)
print(num_int, num_float)

# Multiple assignment
a, b, c = 1, 2, 3
print(a, b, c)
```

```python
# Day 3 – Operators
x = 10
y = 5

# Arithmetic
print(x + y, x - y, x * y, x / y, x % y, x ** y, x // y)

# Comparison
print(x > y, x < y, x == y, x != y)

# Logical
print(x > 5 and y < 10, x > 5 or y > 10, not(x == y))
```

```python
# Day 4 – Strings
text = "My name is Kevin and Python is fun."
print(text.upper())
print(text.lower())
print(text.strip())
print(text.replace("fun", "awesome"))
print(text[0], text[-1])
print(text[0:6])  # slicing
print(text.split())
```

```python
# Day 5 – Lists
fruits = ["apple", "banana", "mango"]
print(fruits[0], fruits[-1])
fruits.append("orange")
fruits.remove("banana")
fruits.insert(1, "kiwi")
print(fruits)
```

```python
# Day 6 – Tuples
student = ("Kevin", 20, "Kenya")
print(student)
name, age, country = student
print(name, age, country)
```

```python
# Day 7 – Sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A.union(B))
print(A.intersection(B))
print(A.difference(B))
```

```python
# Day 8 – Dictionaries
student = {"name": "Kevin", "age": 20}
print(student["name"])
student["course"] = "Data Science"
student["age"] = 20
del student["course"]
print(student)
```

```python
# Day 9 – If, Elif, Else
age = 18

if age < 18:
    print("Minor")
elif age == 18:
    print("Just became adult")
else:
    print("Adult")
```

```python
# For loop
for i in range(1, 6):
    print(i)

# While loop
count = 1
while count <= 5:
    print(count)
    count += 1
```

```python
# Day 11 – Functions
def greet(name):
    return f"Hello, {name}!"

print(greet("Kevin"))

def add(a, b):
    return a + b

print(add(5, 3))
```

```python
# Day 12 – Modules and Packages

# Importing built-in module
import math
print(math.sqrt(16))
print(math.pi)
```

```python
#Day 13 – List Comprehensions

numbers = [1, 2, 3, 4, 5]
squares = [x**2 for x in numbers]
print(squares)
```

```python
#Day 14 – Lambda Functions

double = lambda x: x*2
print(double(5))

nums = [1,2,3,4]
doubled = list(map(lambda x: x*2, nums))
print(doubled)

evens = list(filter(lambda x: x%2==0, nums))
print(evens)
```

```python
#Day 15 - Exception Handling

try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("Execution completed")
```

```python
#Day 20 - Classes and Objects

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p = Person("Kevin", 20)
print(p.name, p.age)
```

```python
#Day 21 - Methods in Classes

class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        return f"{self.name} says Woof!"

d = Dog("Rex")
print(d.bark())
```
